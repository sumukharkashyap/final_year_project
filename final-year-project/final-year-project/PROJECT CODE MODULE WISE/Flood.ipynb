```python
#!pip install seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import os
print(os.listdir())

import warnings
warnings.filterwarnings('ignore')
```

```python
data_1 = pd.read_csv('kerala.csv')
```

```python
data_1
```

```python
data_1.drop(['SUBDIVISION'], axis= 1)
```

```python
data_1.info()
```

```python
data_1['FLOODS'].replace(['YES','NO'],[1,0],inplace=True)
```

```python
data_1
```

```python
x1 = data_1.iloc[:,1:14]
x1.head()
```

```python
y1 = data_1.iloc[:, -1]
y1.head()
```

```python
#Let's see how the rainfall index vary during rainy season
import matplotlib.pyplot as plt
%matplotlib inline
c = data_1[['JUN','JUL','AUG','SEP']]
c.hist()
plt.show()
```

```python
#Data might be widely distributed so let's scale it between 0 and 1
from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x1).transform(x1)
```

```python
#Let's divide the dataset into 2 sets:train and test in ratio (4:1)
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2)
```

```python
x1_train.head()
y1_test.head()
```

```python
clf = neighbors.KNeighborsClassifier()
knn_clf = clf.fit(x1_train,y1_train)
```

```python
#Let's predict chances of flood
y1_predict = knn_clf.predict(x1_test)
print('predicted chances of flood')
print(y1_predict)
```

```python
#Actual chances of flood
print("actual values of floods:")
print(y1_test)
from sklearn.metrics import accuracy_score
pred_score = accuracy_score(y1_test,y1_predict)
pred_score
```
